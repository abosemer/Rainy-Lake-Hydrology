### Rainy Lake Rule Curves

```python
def plotRuleCurve(RC):
    plt.figure(figsize=(12,5))
    plt.hold(True)
    RC['URC'].plot(linewidth=1,color='0.75');
    RC['LRC'].plot(linewidth=1,color='0.75');
    if 'EDL' in RC.columns:
        RC['EDL'].plot(linewidth=3,color='r');
    plt.fill_between(RC.index,RC['LRC'].tolist(),RC['URC'].tolist(),
                     color='0.75')
    plt.hold(False)
    plt.ylabel('level [meters]')
```
