# Changes in Rainy River Flows 1970-2010

```python
# Display graphics inline with the notebook
%matplotlib inline

# Standard Python modules
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import os
import datetime

# Module to enhance matplotlib plotting
import seaborn
seaborn.set()

# Modules to display images and data tables
from IPython.display import Image
from IPython.core.display import display
```

```python
RR = pd.read_pickle('RR.pkl')

plt.subplot(2,1,1)
RR.plot()
plt.title(RR.name)
plt.ylabel('Flow [cubic meters/sec]');

plt.subplot(2,1,2)
RR.hist(bins=100)
plt.xlabel('Flow [cubic meters/sec]')
plt.ylabel('Count')
```

### Comparing Mean Annual Flows, 1970-2000 and 2000-2010

```python
Qmean = [RR[RR.index.year == yr].mean() for yr in range(1970,2013)]

Qmean = pd.DataFrame(Qmean,index=range(1970,2013),columns=['Annual Mean Flow'])
Qmean.plot(kind='bar')
plt.title('Annual Mean Flow')
plt.ylabel('[cubic meters/sec')

print Qmean.ix[1970:2000].mean()
plt.hold(True)
plt.plot([1970,2000],[Qmean.ix[1970:2000].mean(),Qmean.ix[1970:2000].mean()])
plt.plot([2000,2013],[Qmean.ix[2000:].mean(),Qmean.ix[2000:].mean()])
plt.hold(False)
print Qmean.ix[2000:].mean()
```

The key observation is that the annual mean flow through upper Rainy River was virtually identical during the period 1970-2000 and 2000-2010. But let's look at the distribution of daily flows.

```python
plt.figure(figsize=(9,8))
plt.subplot(3,1,1)
plt.hold(True)
RR['1970':'2000'].plot(color='b')
RR['2000':].plot(color='r')
plt.hold(False)
plt.title('Daily Flow Rate in Upper Rainy River')
plt.ylabel('cubic meters/sec')
plt.subplot(3,1,2)
RR['1970':'2000'].hist(bins=np.arange(0,1200,20),color='b')
plt.xlabel('cubic meters/sec')
plt.ylabel('Days')
plt.title('Distribution of Daily Flows 1970-2000')
plt.subplot(3,1,3)
RR['2000':].hist(bins=np.arange(0,1200,20),color='r')
plt.xlabel('cubic meters/sec')
plt.ylabel('Days')
plt.title('Distribution of Daily Flows 2000-2010')
plt.tight_layout()
```

### Stage-Frequency Flow Relationship for Upper Rainy River

```python
plt.hold(True)

hist,bins = np.histogram([q for q in RR['1970':'2000'] if pd.notnull(q)],
                         bins = 100)
c = np.cumsum(hist)
c = [1-float(x)/c.max() for x in c]
plt.semilogx(c,bins[:-1])

hist,bins = np.histogram([q for q in RR['2000':] if pd.notnull(q)],
                         bins = 100)
c = np.cumsum(hist)
c = [1-float(x)/c.max() for x in c]
plt.semilogx(c,bins[:-1])
plt.gca().invert_xaxis()
plt.legend(['1970-2000','2000-2010'])

plt.xlabel('Probability to Exceed a Given Flowrate')
plt.ylabel('Flow [cubic meters/sec')
plt.title('Stage-Frequency Diagram for Flows on Upper Rainy River')

plt.hold(False)
```
