# Rule Curves for Rainy and Namakan Lakes

[Rule curves for Rainy and Namakan Lakes were first implemented in 1949](http://www.ijc.org/files/tinymce/uploaded/1949-06-08_IJC_Order.pdf) by the International Joint Commission. The authority for establishing the rule curves is contained in Article 1 of a binational convention signed in Ottawa in 1938:

> The International Joint Commission, established pursuant to the 
> provisions of the treaty signed at Washington on the 11th day of
> January, 1909, relating to questions arising between the United
> States of America and Canada, is hereby clothed with power to 
> determine when emergency conditions exist in the Rainy Lake 
> watershed, whether by reason of high or low water, and the 
> Commission is hereby empowered to adopt such measures of control
> as to it may seem proper with respect to existing dams at Kettle
> and International Falls, as well as with respect to any existing
> or future dams or works in the boundary waters of the Rainy Lake
> watershed, in the event the Commission shall determine that such
> emergency conditions exist.

The rules curves established operating rules for the dams at Kettle
and International Falls for the purpose of avoiding emergency high
and low  water conditions. The documentation established a working
compromise between the competing interests of commercial power 
generation, of the Minnesota and Ontario governments, and of riparian
owners and properietors.

High water events in 1950 and 1954, and complaints lodged with the IJC,
led to a [revision of the rule curves for Namakan Lake in 1957](http://www.ijc.org/files/tinymce/uploaded/1957-10-01_IJC_Order.pdf). The 
main goal was to establish flexibility in operations to accomodate 
situations where unusually large Spring runoff were anticipated:

> It would be beneficial to the various interests 
> if the operating levels of Namakan Lake were modified 
> setting a maximum to which the levels may be raised during 
> the period 1 October to 1 June inclusive, rather than a 
> mandatory, precise level for that period. These maximum 
> allowable levels would be used in conjunction with the 
> levels now specified by rule curve thereby allowing a range 
> of levels between the present rule curve and the proposed 
> maximum levels. 

High and lower water events in the 1960's led to further changes in the
rule curves. The [rule curves of 1970](http://www.ijc.org/files/tinymce/uploaded/1970-07-29_IJC_Order.pdf) established high and low water level and flow conditions for Namakan and Rainy Lakes. The principle changes were to establish a range of acceptable lake levels, and emergency low level and outflow conditions.

## <a name=section01></a>1970 Rule Curves for Rainy and Namakan Lakes

[1970 Rule Curve](http://www.ijc.org/files/tinymce/uploaded/1970-07-29_IJC_Order.pdf)

```python
# Generic Year to represent data
yr = 2014

# Rainy Lake
RL1970  = {
    datetime.datetime(yr, 1, 1): [1106.6, 1107.1],
    datetime.datetime(yr, 2, 1): [1105.8, 1106.6],
    datetime.datetime(yr, 3, 1): [1105.2, 1106.2],
    datetime.datetime(yr, 4, 1): [1104.6, 1105.6],
    datetime.datetime(yr, 4,21): [1104.6, 1106.2],
    datetime.datetime(yr, 5, 1): [1105.1, 1106.6],
    datetime.datetime(yr, 6, 1): [1106.6, 1107.6],
    datetime.datetime(yr, 7, 1): [1107.4, 1108.1],
    datetime.datetime(yr, 8, 1): [1107.4, 1108.1],
    datetime.datetime(yr, 9, 1): [1107.4, 1108.1],
    datetime.datetime(yr,10, 1): [1107.4, 1108.1],
    datetime.datetime(yr,10,11): [1107.4, 1108.1],
    datetime.datetime(yr,11, 1): [1107.2, 1108.1],
    datetime.datetime(yr,12, 1): [1106.8, 1107.6],
    datetime.datetime(yr,12,31): [1106.4, 1107.1]
}

RL1970 = pd.DataFrame.from_dict(RL1970).transpose()/3.2808
RL1970.columns = ['LRC','URC']
RL1970 = RL1970.resample(rule='1D').interpolate()

# Namakan Lake
NL1970 = {    
    datetime.datetime(yr, 1, 1): [1113.6, 1115.3],
    datetime.datetime(yr, 2, 1): [1111.9, 1114.1],
    datetime.datetime(yr, 3, 1): [1110.3, 1113.1],
    datetime.datetime(yr, 4, 1): [1108.6, 1112.0],
    datetime.datetime(yr, 4,21): [1108.6, 1113.1],
    datetime.datetime(yr, 5, 1): [1110.2, 1113.6],
    datetime.datetime(yr, 6, 1): [1115.3, 1116.6],
    datetime.datetime(yr, 6,21): [1117.5, 1118.6],
    datetime.datetime(yr, 7, 1): [1117.6, 1118.6],
    datetime.datetime(yr, 7,21): [1118.0, 1118.6],
    datetime.datetime(yr, 8, 1): [1118.0, 1118.6],
    datetime.datetime(yr, 9, 1): [1118.0, 1118.6],
    datetime.datetime(yr, 9,11): [1118.0, 1118.6],
    datetime.datetime(yr,10, 1): [1117.6, 1118.6],
    datetime.datetime(yr,11, 1): [1116.3, 1117.5],
    datetime.datetime(yr,12, 1): [1115.0, 1116.5],
    datetime.datetime(yr,12,31): [1113.6, 1115.3]
}

NL1970 = pd.DataFrame.from_dict(NL1970).transpose()/3.2808
NL1970.columns = ['LRC','URC']
NL1970 = NL1970.resample(rule='1D').interpolate()

# Plot 1970 Rule Curves

plt.fill_between(NL1970.index,
                 NL1970['LRC'].tolist(),
                 NL1970['URC'].tolist(),
                 color='y',
                 alpha='1')

plt.fill_between(RL1970.index,
                 RL1970['LRC'].tolist(),
                 RL1970['URC'].tolist(),
                 color='y',
                 alpha='1')

plt.ylabel('Lake Level [meters]')
plt.title('Rule Curves for Namakan and Rainy Lakes: 1970-2000')

plt.tight_layout()
```
