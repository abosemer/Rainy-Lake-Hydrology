## Water Survey of Canada

Historical data for lake levels and stream flows in the Rainy Lake watershed are publically available from several sources include the [Water Survey of Canada](http://wateroffice.ec.gc.ca/).

For the calculations in these notebooks, the [HYDAT database](https://www.ec.gc.ca/rhc-wsc/default.asp?lang=En&n=9018B5EC-1) containing historical data from the water survey of Canada was downloaded from Environment Canada. The database consists of a Microsoft Access .mdb file. Individual tables from the database were extracted to .csv files using [MDB Explorer](http://www.mdbexplorer.com/) which are located and imported in the following cell.

### Read Data Tables

```python
# Directory where data files have been stored
dir = "./Data/"

# The following csv data tables were extracted from the HYDAT .mdb dataset
STATIONS = pd.read_csv(dir + 'STATIONS.csv', index_col = 0);
DLY_LEVELS = pd.read_csv(dir + 'DLY_LEVELS.csv');
DLY_FLOWS = pd.read_csv(dir + 'DLY_FLOWS.csv');
```

## Mapping WSC Stations

The HYDAT database is a collection of data associated with monitoring stations located throughout Canada. The STATIONS table contains a list of stations and attributes, including the latitude and longitude of their position. As an example, here we list attributes for 05PB007, the station that monitors the level of Rainy Lake near Fort Frances, Ontario.

```python
STATIONS.ix['05PB007']
```

The following function maps a list of stations identified by their station numbers. In extracts latitude and longitude from the STATIONS table, then calls the google maps web api to create a map image.

```python
def mapWSC(stationList):
    S = STATIONS.ix[stationList,['STATION_NAME','LATITUDE','LONGITUDE']]
    locs = ["{0},{1}".format(S.ix[s,'LATITUDE'], \
                              S.ix[s,'LONGITUDE']) \
             for s in S.index]
    google_maps_url = \
        "https://maps.googleapis.com/maps/api/staticmap?" + \
        "size=640x320" + \
        "&maptype=terrain" + \
        "&markers=color:red%7Csize:mid%7C" + "|".join(locs)
    display(Image(url = google_maps_url))
    return S
```

### Level Monitoring Stations

```python
levelStations = ['05PB007','05PB024','05PB018','05PA006','05PA013', \
               '05PB023','05PC024','05PC025','05PA011','05PA003']

mapWSC(levelStations)
```

### Flow Monitoring Stations

```python
flowStations = ['05PC018','05PC019','05PB014']
mapWSC(flowStations)
```

## Reading Daily Level Records

The next cell defines a utility function to create Pandas Series object containing a history of levels data for a particular sensor in the WSC data. The returned object indexed with a daily timestamp.

```python
def getLevelsWSC(s):
    data = DLY_LEVELS[DLY_LEVELS['STATION_NUMBER'] == s]
    ts = {}
    for k in data.index:
        mo = str(data.ix[k,'MONTH'])
        yr = str(data.ix[k,'YEAR'])
        for n in range(1,data.ix[k,'NO_DAYS']+1):
            ts[pd.to_datetime(mo+'/'+str(n)+'/'+yr)] = data.ix[k,'LEVEL'+str(n)]  
    ts = pd.Series(ts)
    ts.name = STATIONS.ix[s,'STATION_NAME'] + ' (' + s + ')'
    # drop any initial null entries
    while pd.isnull(ts.ix[0]):
        ts = ts[1:]
    return ts
```

Use of this function is demonstrated by reading and plotting the history of lake levels for Rainy and Namakan Lakes.

```python
RL = getLevelsWSC('05PB007')
NL = getLevelsWSC('05PA003')
plt.figure(figsize=(9,7))
plt.hold(True)
RL.plot()
NL.plot()
plt.hold(False)
plt.legend([RL.name,NL.name]);
```

## Reading Daily Flow Records

The next cell defines a utility function to create Pandas Series object containing a history of flow data for a particular sensor in the WSC data. The returned object indexed with a daily timestamp.

```python
def getFlowsWSC(s):
    data = DLY_FLOWS[DLY_FLOWS['STATION_NUMBER'] == s]
    ts = {}
    for k in data.index:
        mo = str(data.ix[k,'MONTH'])
        yr = str(data.ix[k,'YEAR'])
        for n in range(1,data.ix[k,'NO_DAYS']+1):
            ts[pd.to_datetime(mo+'/'+str(n)+'/'+yr)] = data.ix[k,'FLOW'+str(n)]  
    ts = pd.Series(ts)
    ts.name = STATIONS.ix[s,'STATION_NAME'] + ' (' + s + ')'
    # drop any initial null entries
    while pd.isnull(ts.ix[0]):
        ts = ts[1:]
    return ts
```

The use of this function is demonstrated by creating a historgram of flows on Rainy River in the period 

```python
q = getFlowsWSC('05PC019')
plt.subplot(2,1,1)
q.plot()
plt.title(q.name)
plt.subplot(2,1,2)
q.hist(bins=100)
```
