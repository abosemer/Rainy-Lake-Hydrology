### Utility Functions

The next cell defines a utility function for extracting and displaying the location of level and flow monitoring stations from the WSC data. 

```python
def mapWSC(stationList):
    S = STATIONS.ix[stationList,['STATION_NAME','LATITUDE','LONGITUDE']]
    locs = ["{0},{1}".format(S.ix[s,'LATITUDE'], \
                              S.ix[s,'LONGITUDE']) \
             for s in S.index]
    google_maps_url = \
        "https://maps.googleapis.com/maps/api/staticmap?" + \
        "size=640x320" + \
        "&maptype=terrain" + \
        "&markers=color:red%7Csize:mid%7C" + "|".join(locs)
    display(Image(url = google_maps_url))
    return S
```

The next cell defines a utility function to create Pandas Series object containing a history of levels data for a particular sensor in the WSC data. The returned object indexed with a daily timestamp.

```python
def getLevelsWSC(s):
    data = DLY_LEVELS[DLY_LEVELS['STATION_NUMBER'] == s]
    ts = {}
    for k in data.index:
        mo = str(data.ix[k,'MONTH'])
        yr = str(data.ix[k,'YEAR'])
        for n in range(1,data.ix[k,'NO_DAYS']+1):
            ts[pd.to_datetime(mo+'/'+str(n)+'/'+yr)] = data.ix[k,'LEVEL'+str(n)]  
    ts = pd.Series(ts)
    ts.name = STATIONS.ix[s,'STATION_NAME'] + ' (' + s + ')'
    # drop any initial null entries
    while pd.isnull(ts.ix[0]):
        ts = ts[1:]
    return ts
```

```python
h = getLevelsWSC('05PB007')
h.plot()
plt.title(h.name)
```

The next cell defines a utility function to create Pandas Series object containing a history of flow data for a particular sensor in the WSC data. The returned object indexed with a daily timestamp.

```python
def getFlowsWSC(s):
    data = DLY_FLOWS[DLY_FLOWS['STATION_NUMBER'] == s]
    ts = {}
    for k in data.index:
        mo = str(data.ix[k,'MONTH'])
        yr = str(data.ix[k,'YEAR'])
        for n in range(1,data.ix[k,'NO_DAYS']+1):
            ts[pd.to_datetime(mo+'/'+str(n)+'/'+yr)] = data.ix[k,'FLOW'+str(n)]  
    ts = pd.Series(ts)
    ts.name = STATIONS.ix[s,'STATION_NAME'] + ' (' + s + ')'
    # drop any initial null entries
    while pd.isnull(ts.ix[0]):
        ts = ts[1:]
    return ts
```

The initial decade of operation of the level sensor at Bear's Pass (05PB024) shows a systematic deviation from the historical data near Fort Frances. Let's look at this more carefully.

```python
plt.subplot(2,1,1)
RL1['1987':].plot()
RL2['1987':].plot()
plt.legend([RL1.name,RL2.name]);
plt.xlabel('Year')
plt.ylabel('Level [meters]')
plt.title('History of Measured Rainy Lake Levels');

plt.subplot(2,1,2)
dR = RL1['1987':] - RL2['1987':]
dR.plot()
plt.xlabel('Year')
plt.ylabel('Level [meters]')
plt.title('Difference in Measured Rainy Lake Levels');

plt.tight_layout()
```

The abrupt change at the beginning of 2009 appears to be a recalibration of the Bear Pass level sensor. To reconcile the level measurements, we find the mean level difference pre- and post- the change in 2009, then add that difference to the levels reported from Bear Pass prior to 2009.

```python
t  = (dR - dR.tshift(1)).argmin()
preAdj = dR[:t][:-1].mean()
postAdj = dR[t:][1:].mean()

RL2[:t][:-1] += preAdj - postAdj

RL1['1987':].plot()
RL2['1987':].plot()
plt.legend([RL1.name,RL2.name]);
plt.xlabel('Year')
plt.ylabel('Level [meters]')
plt.title('History of Measured Rainy Lake Levels');
```

### Statistical Properties of the Level Sensors

The availability of two sensors provides an opportunity to estimate the distribution of the level measurement noise.

```python
dR = RL2 - RL1
dR = dR[dR.notnull()]
plt.subplot(2,1,1)
dR.plot()

plt.subplot(2,1,2)
dR.hist(bins = round((0.5*dR.size)**0.5),normed=1)
mu = dR.mean();
sigma = dR.std();
plt.title('Histogram of Differences between Level Measurements')

ax = plt.axis()
r = np.linspace(ax[0],ax[1],100)
p = [np.exp(-0.5*((x-mu)/sigma)**2)/(sigma*(2*np.pi)**0.5) for x in r]
plt.plot(r,p,linewidth=3)
plt.tight_layout()
```
